```python
from PIL import Image #Biblioteca PIL para procesar imagenes
import matplotlib.pyplot as plt
import numpy as np
```

```python
#leer  una imagen y guardarla en la variable img
# el  argumento "L" establece la imagen como escala de grises
img = Image.open("color2.png").convert("L")
```

```python
#convertir objeto de PIL  a un arreglo de numpy
img=np.array(img)
```

```python
#Convertir arreglo ndarray a imagen  PIL
img2 = Image.fromarray(img)
```

```python
#guardar archivo
img2.save("grayscale.png")
```

```python
#mostrar resultado
plt.imshow(img,"gray")
plt.show()
```
