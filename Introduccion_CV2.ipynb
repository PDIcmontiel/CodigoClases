```python
import cv2 #Biblioteca CV2 para procesar la imagenes
import matplotlib.pyplot as plt 
```

```python
print (cv2.__version__)
```

```python
#Leer una imagen  y convertir   a un  ndarray
img = cv2.imread("color.png")
```

```python
# Convertir la imagen  a escala de grises
img_grayscale = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
#plt muestrara la imagen.
#El primer argumento es la imagen que se mostrara
#El segundo argumento impone que la imagen se muestre en  gris
plt.imshow(img_grayscale,"gray")
```

```python
#guarda  la imagen con  el  primer argumento como nombre de la imagen
#el segundo argumento es la imagen que se guardara
cv2.imwrite("escala_grises.jpg",img_grayscale) 
```
